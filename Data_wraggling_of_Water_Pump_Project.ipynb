```python
# Status_group column has values "functional","non functional" and "functional needs repair"
df['status_group'] = df['status_group'].map({'functional': 0, 'non functional': 1, 'functional needs repair': 2})
```

```python
df['status_group'] = df['status_group'].astype('category')

# Categorical summary: counts and percentages
counts = df['status_group'].value_counts()
perc   = df['status_group'].value_counts(normalize=True) * 100
```

Analysing numeric data

```python
df_num = df.select_dtypes(include=['number']).copy()
df_num.columns
```

```python
df_num.drop(columns=['id'], inplace=True)
```

```python
df_num.describe().T
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
for col in ['longitude', 'latitude']:
    df[col] = df[col].replace(0, np.nan)
    df[col] = df[col].fillna(df[col].median())
```

Resolving Skew variables

```python
df_num['amount_tsh_log'] = np.log1p(df_num['amount_tsh'])
df_num['population_log'] = np.log1p(df_num['population'])

# Handle negative values for gps_height and latitude before log1p transformation
# Add the absolute minimum value to shift the data to be non-negative
df_num['gps_height_log'] = np.log1p(df_num['gps_height'] + abs(df_num['gps_height'].min()))
df_num['longitude_log'] = np.log1p(df_num['longitude'])
df_num['latitude_log'] = np.log1p(df_num['latitude'] + abs(df_num['latitude'].min()))

df_num['construction_year'] = df_num['construction_year'].replace(0, df_num['construction_year'].median())
df_num['construction_year_log'] = np.log1p(df_num['construction_year'])

df['region_code'] = df['region_code'].astype('category')
df_num['district_code'] = df_num['district_code'].astype('category')
```

```python
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap="RdYlGn", annot_kws={"size":15})
```

Analysing categorical data

```python
categorical_values = []
for column in df.columns:
    print('==============================')
    print(f"{column} : {df[column].unique()}")
    if len(df[column].unique()) <= 10:
        categorical_values.append(column)
```

```python
cat_summary = pd.DataFrame({
    'column': [col for col in df.columns
               if (df[col].dtype == 'object' or str(df[col].dtype) == 'category')],
    'unique_values': [df[col].nunique() for col in df.columns
                      if (df[col].dtype == 'object' or str(df[col].dtype) == 'category')]
})

cat_summary = cat_summary[cat_summary['unique_values'] < 135]
print(cat_summary)
```

```python
plt.figure(figsize=(60, 50))
for i, column in enumerate(categorical_values, 1):
    plt.subplot(10, 5, i)
    sns.barplot(x=f"{column}", y='status_group', data=df)
    plt.ylabel('Possibility of having a Functional Pump')
    plt.xlabel(f'{column}')
```
